```python
import pandas as pd
import numpy as np
```

```python
# 1. Ler o Excel a partir da linha correta
arquivo_excel = 'Brent_1_3_5_7_9_13.xlsx'
```

```python
# A linha com os nomes das colunas de preços (PX_LAST) parece ser a 8, então header=7
df = pd.read_excel(arquivo_excel, header=7)
```

```python
# 1. Remover a primeira coluna ("Unnamed: 0")
df = df.drop(df.columns[0], axis=1)
```

```python
# 2. Renomear a coluna de datas para 'Date'
df = df.rename(columns={df.columns[0]: 'Date'})
df
```

```python
# 3. Renomear as colunas de preços
# Ajuste os nomes conforme o número de colunas de preços no seu arquivo!
novos_nomes = ['1_month', '3_months', '5_months', '7_months', '9_months', '13_months']
colunas_precos = df.columns[1:1+len(novos_nomes)]
renomear_dict = dict(zip(colunas_precos, novos_nomes))
df = df.rename(columns=renomear_dict)
df
```

```python
# 4. Converter 'Date' para datetime e filtrar por data
df_filtrado = df[
    (df['Date'] >= pd.to_datetime('2000-01-03')) &
    (df['Date'] <= pd.to_datetime('2025-03-07'))
]
df_filtrado
```

```python
# 1. Defina a coluna 'Date' como índice (necessário para resample)
df_filtrado = df_filtrado.set_index('Date')
df_filtrado
```

```python
# 3. Calcule a média semanal com semanas terminando na sexta-feira
medias_semanais = df_filtrado.resample('W-FRI').mean()
medias_semanais
```

```python
medias_semanais_log = np.log(medias_semanais)
medias_semanais_log.to_csv('price_log_1_3_5_7_9_13.csv', index=False, header=False)
```
