```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Exemplo: dados históricos (log preços)
# logS = np.log(df['OilPrice'].values)  # Seu vetor real
# time = df['Year'].values              # Seu vetor de tempo real

# Aqui, um exemplo sintético:
years = np.arange(1995, 2016, 1/12)  # mensal
logS = 3 + 0.1 * np.sin(0.5 * years) + 0.1 * np.random.randn(len(years))
```

```python
def ss_model_simulate(params, S0, dt, N, seed=None):
    np.random.seed(seed)
    # params: dicionário com os parâmetros calibrados
    kappa = params['kappa']
    sigma_chi = params['sigma_chi']
    risk_chi = params['risk_chi']
    mu_xi = params['mu_xi']
    sigma_xi = params['sigma_xi']
    risk_xi = params['risk_xi']
    rho = params['rho']
    chi0 = params['chi0']
    xi0 = params['xi0']

    chi = np.zeros(N)
    xi = np.zeros(N)
    S = np.zeros(N)
    chi[0], xi[0] = chi0, xi0
    S[0] = S0

    for t in range(1, N):
        epsilon_xi = np.random.randn()
        epsilon_chi = rho * epsilon_xi + np.sqrt(1 - rho**2) * np.random.randn()
        xi[t] = xi[t-1] + (mu_xi - risk_xi) * dt + sigma_xi * epsilon_xi * np.sqrt(dt)
        chi[t] = chi[t-1] * np.exp(-kappa * dt) \
                 - (1 - np.exp(-kappa * dt)) * (risk_chi / kappa) \
                 + sigma_chi * epsilon_chi * np.sqrt((1 - np.exp(-2 * kappa * dt)) / (2 * kappa))
        S[t] = chi[t] + xi[t]

    return S
```

```python
# Frequências de amostragem em meses
sampling_months = [1, 3, 5, 9, 13, 17]

plt.figure(figsize=(12, 8))
for i, n_months in enumerate(sampling_months, 1):
    step = n_months
    idx = np.arange(0, len(years), step)
    years_sampled = years[idx]
    logS_sampled = logS[idx]
    N = len(years_sampled)
    
    # Simulação do modelo calibrado para mesma frequência
    # params = { ... } # coloque aqui os parâmetros calibrados
    # S_sim = ss_model_simulate(params, logS_sampled[0], dt=n_months/12, N=N)
    # Usando dados reais para visualização
    S_sim = logS_sampled  # Aqui coloque a série modelada/filtrada
    
    plt.subplot(2, 3, i)
    plt.plot(years_sampled, S_sim, 'r-', label='Calibration', linewidth=1)
    plt.plot(years_sampled, logS_sampled, 'ko', label='Data', markersize=3)
    plt.title(f'{n_months} Months')
    if i in [1, 4]: plt.ylabel('log($/BBL)')
    plt.xlabel('Year')
    plt.grid(True, alpha=0.4)
    if i == 2:
        plt.legend(fontsize=9)
plt.suptitle('Oil Price Calibration', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
